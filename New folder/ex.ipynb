```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 6)
```

```python
def exponential_signal(t, A, a):
    return A*(np.e**(a*t))

fig, ax = plt.subplots(2, 1)
plt.suptitle('Exponential Signal', fontsize= 19, weight= 'semibold')

A = 10
a = 1

rng = 5
n = np.arange(-rng, rng +1, 1)
y = exponential_signal(n, A, a)

plt.tight_layout()
ax[0].stem(n, y)
ax[0].set_title('Discrete Exponential Signal')
ax[0].set_xlabel('Time (second)')
ax[0].set_ylabel('Ampitude (voltage)')

t = np.arange(-rng, rng + .01, .01)
y = exponential_signal(t, A, a)

plt.tight_layout()
ax[1].plot(t, y)
ax[1].set_title('Continuous Exponential Signal')
ax[1].set_xlabel('Time (second)')
ax[1].set_ylabel('Ampitude (voltage)')
plt.show()
```
